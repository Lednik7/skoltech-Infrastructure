```python
import os
import numpy as np

from PIL import Image
from matplotlib import pyplot as plt
```

```python
from matplotlib import pyplot as plt

def overlay_mask(image_path, mask_path):
    image = Image.open(image_path)
    mask = Image.open(mask_path)
    
    _, axs = plt.subplots(1, 2, figsize=(12, 12))

    axs[0].imshow(image)
    axs[0].set_title('Исходное изображение')

    axs[1].imshow(mask, cmap=plt.cm.Greys)
    axs[1].set_title('Изображение с маской')

    plt.show()
```

```python
DATA_PATH = '../data/digital_leaders'

path_to_image = os.path.join(DATA_PATH, 'images/train_image_000.png')
path_to_mask = os.path.join(DATA_PATH, 'masks/train_mask_000.png')

overlay_mask(path_to_image, path_to_mask)
```
